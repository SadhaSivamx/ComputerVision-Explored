```python
#initializing libraries
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Converting BGR to RGB
img=cv.imread("BarackObama.jpg")
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
NewImg=[]
# For each Row
for Rele in img:
  Temp=[]
  #Converting each element to grayscale Value
  for ele in Rele:
    Redval,Greenval,Blueval=list(ele)
    Gval=(0.2989*Redval)+(0.5870*Greenval)+(0.1140*Blueval)
    Temp.append(round(Gval))
  NewImg.append(Temp)
plt.imshow(NewImg,cmap="gray")
```
