```python
#Importing
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
#Defining a Function
def Maxpooling(image,Size=10):
  #Reading as numpy array
  Imgarr=cv.imread(image)
  #Grayscale convertion
  GImg=cv.cvtColor(Imgarr,cv.COLOR_BGR2GRAY)
  #Defining Heights,Weights
  h=len(GImg)
  w=len(GImg[0])
  wn,hn=[0,0]
  #Setting New Image
  NewImg=[]
  for i in range(0,h-Size,Size):
    Hele=GImg[i:i+Size]
    #Getting Max-value
    Rele=[]
    for j in range(0,w-Size,Size):
      Rele.append(np.max((Hele[:,j:j+Size])))
    #Adding To NewImg
    NewImg.append(Rele)
  return NewImg
```

```python
ImgArr=cv.imread("Doggy.jpg")
plt.imshow(ImgArr)
```

```python
Img=np.array(Maxpooling("Doggy.jpg",Size=20))
plt.imshow(Img)
```
