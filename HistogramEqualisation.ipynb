```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
img=cv.imread("People.jpeg",0)
S=(img.shape)
plt.imshow(img,cmap="gray")
```

```python
mini = img.min()
maxi = img.max()
i = [0 for _ in range(0, maxi + 1)]
Newimg = img.flatten()

for x in Newimg:
    i[x] += 1

PDF = [count / sum(i) for count in i]

CDF = [PDF[0]]
for m in range(1, len(PDF)):
    CDF.append(PDF[m] + CDF[-1])

# Scale CDF to the original intensity range
CDF = [int(k * maxi) for k in CDF]

Histogram = np.array(CDF, int)

MappedImg = [Histogram[z] for z in Newimg]

MappedImg = np.array(MappedImg)
MappedImg = MappedImg.reshape(S)

plt.imshow(MappedImg, cmap="gray")
plt.show()
```
