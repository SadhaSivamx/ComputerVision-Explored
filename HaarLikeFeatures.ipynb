### Calculation of Integral Image

```python
def IntegralSum(img):
    img = img.astype(np.int32)
    rimg = img.copy()
    for i in range(len(img)):
        for j in range(len(img[0])):
            if j == 0 and i == 0:
                img[i][j] = img[i][j]
            elif j > 0 and i == 0:
                img[i][j] = img[i][j] + img[i][j-1]
            elif i > 0 and j == 0:
                img[i][j] = img[i][j] + img[i-1][j]
            elif i > 0 and j > 0:
                img[i][j] = img[i][j] + img[i-1][j] + img[i][j-1] - img[i-1][j-1]
    return rimg, img
```

### Calculation of Pixel Sum Under a Certain Area

```python
def IntRectSum(i, j, Int, kernel):
    kernel = kernel - 1
    return Int[i + kernel][j + kernel] - Int[i - 1][j + kernel] - Int[i + kernel][j - 1] + Int[i - 1][j - 1]
```

### Demo for Feature Calculation for Horizontal and Vertical Edges



image.png

```python
def CalculateVector(img, kernel=10):
    Vector = []
    _, Integral = IntegralSum(img)
    for i in range(1, len(img) - (kernel * 2) + 1):
        for j in range(1, len(img[0]) - (kernel * 1) + 1):
            Fe1 = IntRectSum(i, j, Integral, kernel)
            Fe2 = IntRectSum(i + kernel, j, Integral, kernel)
            Vec1 = Fe1 - Fe2
            Vector.append(Vec1)

    for i in range(1, len(img) - (kernel * 1) + 1):
        for j in range(1, len(img[0]) - (kernel * 2) + 1):
            Fe3 = IntRectSum(i, j, Integral, kernel)
            Fe4 = IntRectSum(i, j + kernel, Integral, kernel)
            Vec2 = Fe3 - Fe4
            Vector.append(Vec2)

    Vector = np.array(Vector)
```
